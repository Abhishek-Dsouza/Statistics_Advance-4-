## Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would use each type of test.
### Answer:-
The main difference between a t-test and a z-test lies in the assumptions made about the population variance and the sample size.

A t-test is used when the population standard deviation is unknown or when the sample size is small. It is appropriate when working with small samples (typically less than 30) or when the population variance is not known. The t-test uses the sample standard deviation to estimate the population standard deviation.

Example scenario for a t-test: Let's say you want to test whether the mean height of a specific group of individuals is significantly different from the average height of the general population. You collect a sample of 25 individuals from the group and measure their heights. Since the population standard deviation is unknown, you would use a t-test to compare the sample mean to the population mean.

A z-test, on the other hand, is used when the population standard deviation is known or when the sample size is large. It assumes that the data are normally distributed. The z-test uses the known population standard deviation to calculate the test statistic.

Example scenario for a z-test: Suppose you have access to a large dataset of students' test scores from a particular school. You want to determine if the average test score of the students in this school is significantly higher than the national average test score. Since you have a large sample size and you can obtain the population standard deviation (e.g., from previous studies or official data), you can use a z-test to compare the sample mean to the population mean.

## Q2: Differentiate between one-tailed and two-tailed tests.
### Answer:-
In statistical hypothesis testing, the distinction between one-tailed and two-tailed tests relates to the directionality of the hypothesis and the regions of the distribution being considered.

One-tailed test: A one-tailed test, also known as a directional test, is used when the research hypothesis specifies a particular direction of effect. It is designed to detect whether the sample data significantly deviate in one specific direction from the null hypothesis. The critical region is located entirely on one side of the distribution.

Example: Let's say you want to test whether a new drug improves a certain condition. The research hypothesis states that the drug has a positive effect. In this case, a one-tailed test would be appropriate because you are only interested in whether the drug has a beneficial impact. The critical region would be on one side of the distribution, either the right side (upper-tail test) if you expect the drug to increase the effect, or the left side (lower-tail test) if you expect the drug to decrease the effect.

Two-tailed test: A two-tailed test, also known as a non-directional test, is used when the research hypothesis does not specify a particular direction of effect. It is designed to detect whether the sample data significantly differ from the null hypothesis in any direction. The critical region is divided between both sides of the distribution.

Example: Let's consider a scenario where you want to test whether there is a difference in average scores between two groups. The research hypothesis would be that there is a difference, but it does not specify the direction (e.g., one group having higher scores than the other). In this case, a two-tailed test would be appropriate because you are interested in detecting any significant difference, regardless of the direction. The critical region would be divided between both tails of the distribution.

## Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.
### Answer:-
Type 1 error (False Positive): A Type 1 error occurs when the null hypothesis is rejected, even though it is true. In other words, it is the incorrect rejection of a true null hypothesis. The significance level, denoted as alpha (α), represents the probability of committing a Type 1 error.

Example scenario: Let's say a pharmaceutical company is testing a new drug for its effectiveness in treating a certain condition. The null hypothesis (H0) is that the drug has no effect. However, due to random variability or other factors, the study incorrectly concludes that the drug is effective (rejecting the null hypothesis) when, in fact, it is not. This would be a Type 1 error, potentially leading to the adoption of an ineffective or unsafe treatment.

Type 2 error (False Negative): A Type 2 error occurs when the null hypothesis is not rejected, even though it is false. It is the failure to reject a false null hypothesis. The probability of committing a Type 2 error is denoted as beta (β).

Example scenario: Continuing from the previous example, suppose the new drug is genuinely effective in treating the condition, but the study fails to detect its effectiveness (fails to reject the null hypothesis). This would be a Type 2 error, potentially resulting in the rejection of a beneficial treatment.

## Q4: Explain Bayes's theorem with an example.
### Answer:- 
Bayes's theorem is a fundamental concept in probability theory and statistics that allows us to update the probability of an event based on new evidence. It provides a framework for revising our beliefs about the likelihood of an event in light of new information.

The theorem is named after Thomas Bayes, an 18th-century mathematician, and it can be stated as follows:

P(A|B) = (P(B|A) * P(A)) / P(B)

where:

P(A|B) is the posterior probability, which is the probability of event A occurring given that event B has occurred.
P(B|A) is the likelihood, which is the probability of event B occurring given that event A has occurred.
P(A) and P(B) are the prior probabilities, which are the probabilities of event A and event B occurring independently.
To understand Bayes's theorem better, let's consider an example:

Example scenario: Imagine you are a doctor trying to diagnose a rare disease. The disease affects 1 in 10,000 people in the general population, and there is a diagnostic test available that correctly identifies the disease 95% of the time (true positive rate) and produces a false positive result in 2% of healthy individuals (false positive rate).

Given this information, let's say a patient tests positive for the disease. We want to determine the probability that the patient actually has the disease (posterior probability).

Using Bayes's theorem, we can calculate this as follows:

P(Disease|Positive) = (P(Positive|Disease) * P(Disease)) / P(Positive)

P(Disease|Positive) is the posterior probability of having the disease given a positive test result.

P(Positive|Disease) is the likelihood of testing positive given that the patient has the disease (true positive rate, 95%).

P(Disease) is the prior probability of having the disease (1 in 10,000).

P(Positive) is the probability of testing positive, considering both true positives (95% of 1 in 10,000) and false positives (2% of 9,999 in 10,000).

By plugging the values into the equation, we can calculate the posterior probability and determine the likelihood that the patient actually has the disease based on the test result.


Bayes's theorem is a powerful tool for updating probabilities based on new evidence and is widely used in fields such as medical diagnosis, machine learning, and Bayesian statistics. It allows for a more nuanced understanding of probabilities by incorporating prior beliefs and updating them with observed data.

## Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.
### Answer:-
A confidence interval is a range of values that is likely to contain the true value of a population parameter with a certain degree of confidence. In other words, it is a range of values around a sample statistic that is likely to contain the true population parameter value with a certain level of certainty or probability.

To calculate a confidence interval, we need to know the sample mean, sample size, and standard deviation (or standard error) of the sample statistic. The formula for a confidence interval is:

Confidence interval = sample mean ± (z-score * standard error)

where z-score is the number of standard deviations corresponding to the desired level of confidence. For example, for a 95% confidence interval, the z-score would be 1.96.

Here's an example of how to calculate a confidence interval:

Suppose we want to estimate the average height of all people in a city. We take a random sample of 50 people and find that the sample mean height is 170 cm, and the sample standard deviation is 5 cm.

To calculate a 95% confidence interval for the population mean height, we need to first calculate the standard error. The standard error is calculated as:

Standard error = sample standard deviation / sqrt(sample size)

= 5 / sqrt(50)

= 0.71

Next, we can calculate the confidence interval as:

Confidence interval = 170 ± (1.96 * 0.71)

= 168.6 to 171.4 cm

This means that we can be 95% confident that the true population mean height falls within the range of 168.6 to 171.4 cm. We can interpret this as if we repeat this sampling process many times, 95% of the time, the true population mean height will be contained within this interval.

This is because a larger sample size reduces the effect of random sampling error, and thus increases the precision of the estimate.

Assuming a normal distribution and using the formula for the margin of error, we can calculate:

Margin of error = z * sqrt(p_hat * (1 - p_hat) / n)

where z is the critical value for a 95% confidence interval, p_hat is the sample proportion, and n is the sample size.

Using a standard normal table, we find that the critical value for a 95% confidence interval is 1.96. Plugging in the values, we get:

Margin of error = 1.96 * sqrt(0.6 * 0.4 / 100)

= 0.098

This means that we are 95% confident that the true population proportion of adults who support the candidate is between 0.502 and 0.698 (0.6 ± 0.098).

Now, suppose we increase the sample size to 400. Using the same formula, we can calculate the new margin of error as:

Margin of error = 1.96 * sqrt(0.6 * 0.4 / 400)

This means that we are 95% confident that the true population proportion of adults who support the candidate is between 0.551 and 0.649 (0.6 ± 0.049).

z = (x - μ) / σ

where x is the value of the data point, μ is the population mean, and σ is the population standard deviation.

z = (75 - 70) / 5

z = 1

This means that the data point of 75 is 1 standard deviation above the population mean. The z-score tells us how many standard deviations a data point is from the population mean. In this case, the z-score of 1 indicates that the data point is one standard deviation above the mean.

In a normal distribution, a z-score of 0 means the data point is at the mean, a z-score of 1 means the data point is one standard deviation above the mean, and a z-score of -1 means the data point is one standard deviation below the mean.

## Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is significantly effective at a 95% confidence level using a t-test.
### Answer:-
To conduct a hypothesis test to determine if the weight loss drug is significantly effective at a 95% confidence level using a t-test, we need to state our null and alternative hypotheses:

Null hypothesis: The average weight loss for participants taking the new drug is not significantly different from 0. Alternative hypothesis: The average weight loss for participants taking the new drug is significantly greater than 0. We will use a one-sample t-test, since we are comparing the mean weight loss of the sample to a hypothesized population mean of 0.

We can calculate the t-statistic using the formula:

t = (x̄ - μ) / (s / √n)

where x̄ is the sample mean, μ is the hypothesized population mean (0 in this case), s is the sample standard deviation, and n is the sample size.

Plugging in the values, we get:

t = (6 - 0) / (2.5 / √50)

t = 15.39

Using a t-table with 49 degrees of freedom and a significance level of 0.05 (since we want a 95% confidence level), we find the critical t-value to be 1.677. Since our calculated t-value of 15.39 is greater than the critical t-value of 1.677, we can reject the null hypothesis and conclude that the average weight loss for participants taking the new drug is significantly greater than 0 at a 95% confidence level.

Therefore, we can conclude that there is evidence to support the effectiveness of the new weight loss drug.
